```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fair_baselines/fb_results/hardt__preds.csv')
df
```

```python
np.mean(df.loc[df.trial ==2].label == df.loc[df.trial ==2].preds)
```

```python
from fairlearn.metrics import selection_rate, true_positive_rate, false_positive_rate
```

```python
for i in range(10):
    currdf = df.loc[df.trial == i]
    a_df = currdf.loc[currdf.group == 1]
    b_df = currdf.loc[currdf.group == 0]
    print("=== selection rates ===")
    print(selection_rate(a_df.label, a_df.preds))
    print(selection_rate(b_df.label, b_df.preds))
    print("=== true pos. rates ===")
    print(true_positive_rate(a_df.label, a_df.preds))
    print(true_positive_rate(b_df.label, b_df.preds))
    print("=== false pos rates ===")
    print(false_positive_rate(a_df.label, a_df.preds))
    print(false_positive_rate(b_df.label, b_df.preds))
```
